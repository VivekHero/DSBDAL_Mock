```python
fig, axes = plt.subplots(nrows = 3, ncols=1, figsize=(8, 6))

for class_label, i in zip(df["Species"].unique(), range(3)):
  class_data = df[df["Species"] == class_label]
  sns.boxplot(x = class_data["SepalLengthCm"], ax = axes[i])
  axes[i].set(title = class_label)
  print(class_label, " Outliers:")
  
  Q1 = class_data["SepalLengthCm"].quantile(0.25)
  Q3 = class_data["SepalLengthCm"].quantile(0.75)
  IQR = Q3 - Q1
  
  threshold = 1.5
  outliers = class_data[(class_data['SepalLengthCm'] < Q1 - threshold * IQR) | (class_data['SepalLengthCm'] > Q3 + threshold * IQR)]
  print("Count: ", len(outliers))
  print(outliers)
  print("----------------------")
plt.tight_layout()
plt.show()
```

```python
#box plot for PetalWidthCm for all species
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(3, 6))
for class_label,i in zip(df['Species'].unique(), range(3)):
  class_data = df[df['Species'] == class_label]
  sns.boxplot(x=class_data["PetalWidthCm"], ax=axes[i])
  axes[i].set( title=class_label)

  print(class_label, " Outliers:")
  # detecting outliers
  # calculate IQR for column Height
  Q1 = class_data['PetalWidthCm'].quantile(0.25)
  Q3 = class_data['PetalWidthCm'].quantile(0.75)
  IQR = Q3 - Q1

  # identify outliers
  threshold = 1.5
  outliers = class_data[(class_data['PetalWidthCm'] < Q1 - threshold * IQR) | (class_data['PetalWidthCm'] > Q3 + threshold * IQR)]
  print("Count: ", len(outliers))
  print(outliers)
  print("----------------------")

plt.tight_layout()
plt.show()
```
