```python
condition = (pd.isna(df['age'])) & (df['class'] == 'First')
mean_value = df.loc[df['class'] == 'First', 'age'].mean()
df.loc[condition, 'age'] = mean_value

condition = (pd.isna(df["age"])) & (df["class"] == "Second")
mean_value = df.loc[df["class"] == "Second", "age"].mean()
df.loc[condition, "age"] = mean_value

condition = (pd.isna(df["age"])) & (df["class"] == "Third")
mean_value = df.loc[df["class"] == "Third", "age"].mean()
df.loc[condition, "age"] = mean_value
```

```python
plt.figure(figsize=(8, 6))

for class_label in df['class'].unique():
  class_data = df[df['class'] == class_label]
  plt.scatter(class_data['age'], [class_label] * len(class_data))

plt.xlabel('Age')
plt.ylabel('Class')
plt.title('Age distribution by class')
plt.legend()
plt.show()

g = sns.FacetGrid(df, col = 'class')
g.map(plt.hist, 'age', bins = 20)
```

```python
plt.figure(figsize=(8,6))

for class_label in df['class'].unique():
  class_data = df[df['class'] == class_label]
  plt.scatter(class_data['fare'], [class_label] * len(class_data), label = class_label)

plt.xlabel('Fare')
plt.ylabel('Class')
plt.title('Fare distribution by class')
plt.legend()
plt.show()

g = sns.FacetGrid(df, col = 'class')
g.map(plt.hist, 'fare', bins = 20)  
```

```python
#Looking for outliers in Fisrt Class Fares
classdata = df[df['class'] == 'First']
```

```python
sns.boxplot(x = classdata['fare'])
```

```python
Q1 = classdata['fare'].quantile(0.25)
Q3 = classdata['fare'].quantile(0.75)
IQR = Q3 - Q1

threshold = 1.5

outliers = classdata[(classdata['fare'] < (Q1 - threshold * IQR)) | (classdata['fare'] > (Q3 + threshold * IQR))]
outliers
```

```python
# detecting outliers
# calculate IQR for column Height
Q1 = new_df['age'].quantile(0.25)
Q3 = new_df['age'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = new_df[(new_df['age'] < Q1 - threshold * IQR) | (new_df['age'] > Q3 + threshold * IQR)]
outliers
```

```python
#removing outliers
#by trimming
trimmed_new_df = new_df.drop(outliers.index)

sns.boxplot(trimmed_new_df['age'])
```

```python
Q1 = new_df['age'].quantile(0.25)
Q3 = new_df['age'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5

lower_whisker = Q1 - whisker_width * IQR
upper_whisker = Q3 + whisker_width * IQR

capped_new_df = new_df
capped_new_df['age'] = np.where(new_df['age'] > upper_whisker, upper_whisker, np.where(new_df['age'] < lower_whisker, lower_whisker, new_df['age']))
```

```python
sns.boxplot(capped_new_df['age'])
```
